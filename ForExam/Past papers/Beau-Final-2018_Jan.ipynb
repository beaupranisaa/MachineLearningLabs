```python
def ff(x,W,b):
    L = len(W)-1
    a = x
    for l in range(1,L+1):
        z = W[l].T*a+b[l]
        if (l == L):
            a = sigmoid_act(z)
        else:
            a = sigmoid_act(z)
    return a

####MSE
def loss_binary(y, yhat):
    return -((1 - y) * np.log(1 - yhat) + y * np.log(yhat))

def loss_multi(y, yhat):
    return - np.dot(y, np.log(yhat))

def loss_mse(y,yhat):
    return np.power(y-yhat,2)
```
